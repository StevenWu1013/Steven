```python
import neurolab as nl
import numpy as np

# Create train samples
i1 = np.array([[-1.5,-1,-0.5,0.5,1,1.5,1.5,1,0.5,-0.5,-1,-1.5,-1.5]])

t1 = np.array([[-0.5,0,0.5,0.5,0,-0.5,0.5,0,-0.5,-0.5,0,0.5,-0.5]])

input = i1.reshape(13,1)
target = t1.reshape(13,1)
# Create network with 2 layers
net = nl.net.newelm([[-1.5, 1.5]], [10,1], [nl.trans.SatLin(),nl.trans.PureLin()])
# Set initialized functions and init
net.layers[0].initf = nl.init.InitRand([-0.1, 0.1], 'wb')
net.layers[1].initf= nl.init.InitRand([-0.1, 0.1], 'wb')
net.init()
# Train network
error = net.train(input, target, epochs=5000, show=100, goal=0.002,lr=0.003)
# Simulate network
output = net.sim(input)

# Plot result
import pylab as pl
%matplotlib inline

pl.subplot(211)
pl.plot(error)
pl.xlabel('Epoch number')
pl.ylabel('Train error (default MSE)')

pl.subplot(212)
pl.plot(i1.reshape(13),output.reshape(13))
#pl.legend(['net output'],'upper right')
pl.show()
```
